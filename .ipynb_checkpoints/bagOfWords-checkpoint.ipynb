```python
import pandas as pd
df = pd.read_csv('campaigns.csv')
df.head()
```

```python
title = ''
content = ''

for _, data in df.iterrows():
    title += data['title'] + ' '
    content += data['plain_text'][500:] + ' '
```

```python
title
```

```python
content
```

```python
title_preproceed = ''
content_preproceed = ''

for temp in title:
  if (temp.isalpha() or temp == ' '):
    title_preproceed += temp

for temp in content:
  if (temp.isalpha() or temp == ' '):
    content_preproceed += temp

title_preproceed = title_preproceed.lower()[:-1]
content_preproceed = content_preproceed.lower()[:-1]
```

```python
title_preproceed
```

```python
content_preproceed
```

```python
bow_title = {}
bow_content = {}

for word in title_preproceed.split(' '):
  if word in bow_title.keys():
    bow_title[word] += 1
  else:
    bow_title[word] = 1

for word in content_preproceed.split(' '):
  if word in bow_content.keys():
    bow_content[word] += 1
  else:
    bow_content[word] = 1
```

```python
dict(sorted(bow_title.items(), key=lambda item: item[1], reverse=True))
```
